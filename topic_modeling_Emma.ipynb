Reddit Topic Modeling (balanced)

```python
# Colab: install once per session
!pip -q install bertopic[all]==0.16.0 hdbscan umap-learn sentence-transformers gensim

import pandas as pd
import numpy as np
import re

from bertopic import BERTopic
from bertopic.representation import KeyBERTInspired, MaximalMarginalRelevance, PartOfSpeech, ZeroShotClassification
try:
    # Newer BERTopic versions include a Seeded representation
    from bertopic.representation import Seeded  # may not exist on older versions
    HAS_SEEDED = True
except Exception:
    HAS_SEEDED = False

from sentence_transformers import SentenceTransformer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# You already did a drive.mount and os.chdir to your project folder
reddit_df = pd.read_csv('comments_reddit_180d_clean_balanced.csv')

# Basic sanity checks
reddit_df = reddit_df.dropna(subset=['text'])
reddit_df['text'] = reddit_df['text'].astype(str).str.strip()
reddit_df = reddit_df[reddit_df['text'].str.len() > 0]

reddit_df['platform'] = reddit_df['platform'].fillna('reddit')
reddit_df['show_key'] = reddit_df['show_key'].astype(str)
print(reddit_df[['platform','show_key','text']].head())
```

```python
def simple_clean(s: str) -> str:
    s = re.sub(r'http\S+', ' ', s)         # URLs
    s = re.sub(r'[\r\n\t]+', ' ', s)       # newlines/tabs
    s = re.sub(r'\s+', ' ', s)             # extra spaces
    return s.strip()

reddit_df['text_clean'] = reddit_df['text'].apply(simple_clean)
```

```python
# A small, fast model is fine; swap to 'all-MiniLM-L6-v2' if you prefer
embed_model = SentenceTransformer("all-MiniLM-L6-v2")

# Prune very-common words and keep unigrams+bigrams
vectorizer_model = CountVectorizer(
    stop_words='english',
    ngram_range=(1,2),
    min_df=2,  # Reduced min_df
    max_df=0.9  # Increased max_df
)
```

```python
seed_map = {
    "performance_praise": ["vocals", "live", "amazing", "energy", "stage", "crowd"],
    "politics": ["political", "protest", "controversy", "message", "symbolism"],
    "production_issues": ["sound", "mic", "audio", "mix", "lighting", "delay"],
    "dance_and_choreo": ["dance", "choreo", "movement", "routine"],
    "setlist_song_choice": ["setlist", "medley", "song", "track", "mix"],
    "brand_sponsorship": ["sponsor", "ad", "brand", "commercial"],
    "nostalgia_and_memories": ["nostalgia", "remember", "throwback", "classic"],
    "halftime_vs_game": ["halftime", "game", "football", "super bowl"]
}
```

```python
def train_bertopic_for_group(texts,
                             n_neighbors=15,
                             min_cluster_size=20,
                             seed_map=None,
                             topn_words=10):
    # Representation pipeline: POS filtering -> KeyBERTInspired -> MMR (diversify)
    pos_rep = PartOfSpeech("en_core_web_sm", top_n_words=topn_words)
    keybert_rep = KeyBERTInspired()
    mmr_rep = MaximalMarginalRelevance(diversity=0.4)

    reps = [pos_rep, keybert_rep, mmr_rep]

    # If seeded representation is available, add it first so it guides terms.
    rep_model = None
    if seed_map and HAS_SEEDED:
        try:
            rep_model = [Seeded(seed_map)] + reps
        except Exception:
            rep_model = reps
    else:
        rep_model = reps

    topic_model = BERTopic(
        embedding_model=embed_model,
        vectorizer_model=vectorizer_model,
        language="english",
        calculate_probabilities=True,
        min_topic_size=min_cluster_size,
        nr_topics=None,                   # let HDBSCAN decide
        verbose=False,
        representation_model=rep_model
    )

    topics, probs = topic_model.fit_transform(texts)

    # If we could not use Seeded, we can still bias labels via ZeroShotClassification
    if seed_map and not HAS_SEEDED:
        try:
            zsc = ZeroShotClassification(list(seed_map.keys()))
            topic_model.set_representations(zsc)
        except Exception:
            pass  # keep existing representations

    # Build a compact topic->words table
    info = topic_model.get_topic_info()
    # Drop the outlier topic -1 and keep 5–10 topics by size
    info = info[info.Topic != -1].sort_values("Count", ascending=False)
    keep_n = max(5, min(10, len(info)))
    info = info.head(keep_n).copy()

    # Get the top words for each kept topic
    rows = []
    for tid in info.Topic.tolist():
        words = topic_model.get_topic(tid)
        top_words = [w for (w, score) in words[:topn_words]] if words else []
        rows.append((tid, ", ".join(top_words)))
    word_df = pd.DataFrame(rows, columns=["Topic","Top_Words"])

    info = info.merge(word_df, on="Topic", how="left")
    return topic_model, topics, probs, info
```

```python
results = []

# If you want to filter to Reddit only (this file is Reddit), keep it; otherwise remove the query.
for (plat, show), g in reddit_df.groupby(['platform','show_key'], sort=False):
    texts = g['text_clean'].tolist()
    if len(texts) < 100:
        # Skip very small groups; adjust threshold if needed
        print(f"Skipping {plat} × {show} (only {len(texts)} docs)")
        continue

    print("="*90)
    print(f"Training BERTopic: platform={plat} | show={show} | docs={len(texts)}")
    model, t_ids, t_probs, info = train_bertopic_for_group(
        texts,
        n_neighbors=15,
        min_cluster_size=20,
        seed_map=seed_map,        # set to None to disable seeding
        topn_words=10
    )

    # Optional: add human-friendly labels now (you can edit later)
    # We map by best-matching seed based on overlap with Top_Words
    def auto_label(row):
        words = set(row['Top_Words'].split(', '))
        best_label, best_overlap = None, 0
        for label, seeds in seed_map.items():
            overlap = len(words.intersection(set(seeds)))
            if overlap > best_overlap:
                best_overlap, best_label = overlap, label
        return best_label if best_label else f"topic_{int(row['Topic'])}"

    info['auto_label'] = info.apply(auto_label, axis=1)

    # Print top words per topic
    for _, r in info.iterrows():
        print(f"[{r['Topic']:>3}] {r['auto_label']:<24} | n={int(r['Count'])} | words: {r['Top_Words']}")

    # Collect for a master summary
    info['platform'] = plat
    info['show_key'] = show
    results.append(info)

topics_summary = pd.concat(results, ignore_index=True) if results else pd.DataFrame()
print("\nDone. Summary columns:", list(topics_summary.columns))
```

```python
if not topics_summary.empty:
    out_path = "reddit_bertopic_summary.csv"
    topics_summary.to_csv(out_path, index=False)
    print(f"Saved: {out_path}")
topics_summary.head(10)
```

```python
from gensim import corpora
from gensim.models.ldamodel import LdaModel

def lda_for_show(df, show_key, num_topics=8, passes=5, topn=10):
    subset = df[df['show_key'] == show_key]
    texts = subset['text_clean'].tolist()
    if len(texts) < 100:
        print(f"Too few docs for LDA: {show_key} ({len(texts)})")
        return

    # Simple tokenization
    tokenized = [re.findall(r"[a-zA-Z']{3,}", t.lower()) for t in texts]

    # Dictionary + corpus
    dictionary = corpora.Dictionary(tokenized)
    dictionary.filter_extremes(no_below=5, no_above=0.6)
    corpus = [dictionary.doc2bow(doc) for doc in tokenized]

    lda = LdaModel(corpus=corpus, id2word=dictionary,
                   num_topics=num_topics, passes=passes, random_state=42)

    for i in range(num_topics):
        terms = lda.show_topic(i, topn=topn)
        words = ", ".join([w for w, p in terms])
        print(f"LDA Topic {i}: {words}")

# Example:
# lda_for_show(reddit_df, show_key="rihanna_2023", num_topics=8)
```

```python
# Example manual edits after you review prints:
manual_labels = {
    # topic_id : "your_label"
    # e.g., 3: "production_issues", 7: "performance_praise"
}

if not topics_summary.empty and manual_labels:
    topics_summary['manual_label'] = topics_summary['Topic'].map(manual_labels).fillna(topics_summary['auto_label'])
else:
    topics_summary['manual_label'] = topics_summary.get('auto_label', topics_summary.get('Name', ''))

topics_summary[['platform','show_key','Topic','Count','manual_label','Top_Words']].sort_values(['platform','show_key','Count'], ascending=[True,True,False]).head(30)
```

Reddit Temporal trend

```python
# ================= ONE-CELL: Earlier 3 vs Last 3 — Relative-to-Halftime Sentiment =================
!pip -q install vaderSentiment

import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# --------- config you can tweak ----------
CSV_PATH = "comments_reddit_180d_clean_balanced.csv"
USE_EASTERN = False        # True -> convert to US/Eastern; False -> keep UTC
REL_WINDOW_HOURS = 12      # window around halftime (±); will auto-widen if sparse (up to ±96h)
SMOOTH_WINDOW = 3          # rolling mean window (hours) for smoothing lines
# -----------------------------------------

# 1) Load + basic clean
df = pd.read_csv(CSV_PATH)
if 'text' not in df.columns:
    raise ValueError("Expected a 'text' column in the CSV.")
df = df.dropna(subset=['text']).copy()
df['text'] = df['text'].astype(str).str.strip()

# 2) Detect and parse timestamp (tz-aware)
TS_CANDIDATES = ['created_dt','created_utc','timestamp','created','created_at']
ts_col = next((c for c in TS_CANDIDATES if c in df.columns), None)
if ts_col is None:
    raise ValueError(f"Could not find a timestamp column among {TS_CANDIDATES}. Please tell me the correct column name.")

if USE_EASTERN:
    created_dt = pd.to_datetime(df[ts_col], errors='coerce', utc=True).dt.tz_convert('US/Eastern')
    tz = 'US/Eastern'
else:
    created_dt = pd.to_datetime(df[ts_col], errors='coerce', utc=True)
    tz = 'UTC'

df = df.assign(created_dt=created_dt).dropna(subset=['created_dt'])
df['hour_bin'] = df['created_dt'].dt.floor('H')

# 3) Sentiment (VADER)
if 'sentiment' not in df.columns:
    analyzer = SentimentIntensityAnalyzer()
    df['sentiment'] = df['text'].apply(lambda x: analyzer.polarity_scores(x)['compound'])

# 4) Try to get a show YEAR per row
#    Priority: year/show_year column -> year parsed from show_key suffix -> infer by nearest halftime within ±60 days
supported_years = [2013, 2014, 2015, 2016, 2017, 2020, 2022, 2023]
halftime_by_year = {
    2013: pd.Timestamp('2013-02-03 20:00:00', tz=tz),
    2014: pd.Timestamp('2014-02-02 20:00:00', tz=tz),
    2015: pd.Timestamp('2015-02-01 20:00:00', tz=tz),
    2016: pd.Timestamp('2016-02-07 20:00:00', tz=tz),
    2017: pd.Timestamp('2017-02-05 20:00:00', tz=tz),
    2020: pd.Timestamp('2020-02-02 20:00:00', tz=tz),
    2022: pd.Timestamp('2022-02-13 20:00:00', tz=tz),
    2023: pd.Timestamp('2023-02-12 20:00:00', tz=tz),
}

# start with NaN year
df['event_year'] = np.nan

# a) explicit year/show_year columns
for cand in ['year','show_year']:
    if cand in df.columns:
        tmp = pd.to_numeric(df[cand], errors='coerce')
        df.loc[tmp.isin(supported_years), 'event_year'] = tmp[tmp.isin(supported_years)]

# b) parse year from show_key suffix like "..._2017"
if df['event_year'].isna().any() and 'show_key' in df.columns:
    m = df['show_key'].astype(str).str.extract(r'_(\d{4})$', expand=False)
    y = pd.to_numeric(m, errors='coerce')
    df.loc[df['event_year'].isna() & y.isin(supported_years), 'event_year'] = y

# c) infer from nearest halftime within ±60 days
def infer_year_from_time(ts):
    best_year, best_diff = None, pd.Timedelta(days=9999)
    for y, ht in halftime_by_year.items():
        d = abs(ts - ht)
        if d < best_diff:
            best_year, best_diff = y, d
    return best_year if best_diff <= pd.Timedelta(days=60) else np.nan

mask_missing = df['event_year'].isna()
if mask_missing.any():
    df.loc[mask_missing, 'event_year'] = df.loc[mask_missing, 'created_dt'].apply(infer_year_from_time)

df = df[pd.notna(df['event_year'])].copy()
df['event_year'] = df['event_year'].astype(int)

# 5) Map halftime by row, compute relative hours
df['halftime'] = df['event_year'].map(halftime_by_year)
delta = df['hour_bin'] - df['halftime']
df['rel_hours'] = delta.dt.total_seconds() / 3600.0

# 6) Choose cohorts: earliest 3 years present vs latest 3 years present
years_present = sorted(df['event_year'].unique().tolist())
if len(years_present) < 2:
    raise ValueError(f"Not enough distinct show years found (have: {years_present}).")

earliest_n = min(3, max(1, len(years_present)//2))   # if fewer shows, split roughly in half
latest_n   = min(3, len(years_present) - earliest_n)

early_years = years_present[:earliest_n]
late_years  = years_present[-latest_n:] if latest_n > 0 else []

# ensure no overlap if total < 6
early_set, late_set = set(early_years), set(late_years) - set(early_years)

def cohort_label(y):
    if y in early_set:
        return f"Earlier {len(early_set)}"
    elif y in late_set:
        return f"Recent {len(late_set)}"
    else:
        return "Other"

df['cohort'] = df['event_year'].apply(cohort_label)

# Keep only the two cohorts
df = df[df['cohort'].isin([f"Earlier {len(early_set)}", f"Recent {len(late_set)}"])].copy()
if df.empty:
    raise ValueError("After cohorting, no rows remain. Check that event_year inference is working.")

# 7) Build trend per cohort (relative hours), widen window if sparse
def build_trend(window_hours):
    rel = df[df['rel_hours'].between(-window_hours, window_hours)].copy()
    if rel.empty:
        return rel, pd.DataFrame()
    rel['rel_hour'] = rel['rel_hours'].round().astype(int)
    trend = (rel.groupby(['cohort','rel_hour'])
               .agg(avg_sent=('sentiment','mean'), n=('sentiment','size'))
               .reset_index()
               .sort_values(['cohort','rel_hour']))
    if trend.empty:
        return rel, trend
    # smooth within each cohort series
    trend['avg_sent_smooth'] = (trend.groupby('cohort')['avg_sent']
                                .transform(lambda s: s.rolling(SMOOTH_WINDOW, center=True, min_periods=1).mean()))
    return rel, trend

window = REL_WINDOW_HOURS
rel_df, trend = build_trend(window)
while (trend.empty or (trend.groupby('cohort').size().lt(3).any())) and window < 96:
    window += 12
    rel_df, trend = build_trend(window)

print(f"Cohorts: early years={sorted(list(early_set))}, recent years={sorted(list(late_set))}")
print(f"Using relative window ±{window} hours")
if not trend.empty:
    print(trend.groupby('cohort').size().rename('points').to_frame(), "\n")

# 8) Plot — one chart, two lines (earlier vs recent)
plt.figure(figsize=(12,6))
if trend.empty:
    # fallback: daily absolute-time by cohort
    print("⚠️ Not enough data near halftime; showing daily fallback by cohort.")
    daily = (df.assign(day=df['created_dt'].dt.floor('D'))
               .groupby(['cohort','day'])
               .agg(avg_sent=('sentiment','mean'), n=('sentiment','size'))
               .reset_index()
               .sort_values(['cohort','day']))
    for cohort, g in daily.groupby('cohort'):
        plt.plot(g['day'], g['avg_sent'], label=cohort)
    plt.axhline(0, linewidth=0.8)
    plt.title("Average Sentiment Over Time (Daily fallback by cohort)")
    plt.xlabel("Date")
    plt.ylabel("Average Sentiment (VADER compound)")
else:
    for cohort, g in trend.groupby('cohort'):
        plt.plot(g['rel_hour'], g['avg_sent_smooth'], label=cohort, linewidth=2)
    plt.axvline(0, linestyle='--')  # halftime
    plt.axhline(0, linewidth=0.8)
    plt.title("Average Sentiment Around Halftime — Earlier 3 vs Recent 3")
    plt.xlabel("Hours relative to halftime (0 = halftime)")
    plt.ylabel("Average Sentiment (VADER compound)")

plt.legend(title="Cohort", loc='best')
plt.tight_layout()
plt.show()

# 9) Quick table for your write-up
if not trend.empty:
    summary_tbl = (trend.groupby('cohort')
                         .agg(mean_sent_overall=('avg_sent','mean'),
                              median_sent_overall=('avg_sent','median'),
                              points=('avg_sent','size'))
                         .reset_index())
    print("\nCohort summary:")
    print(summary_tbl.to_string(index=False))
# ================================================================================================
```
