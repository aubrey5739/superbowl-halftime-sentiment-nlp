```python
# ============================================
# Roc Nation era comparison: sentiment, engagement, topics
# ============================================




# --- 0) Setup
import warnings; warnings.filterwarnings("ignore")
import os, glob, re
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import statsmodels.stats.api as sms
```

```python
# --- 1) Load data
# Adjust these to your paths if needed.
# We'll try to auto-find the files used in your notebook by pattern.
candidates = [
    "/content/comments_reddit_180d_clean_balanced.csv",
    "/content/comments_yt_180d_clean_balanced.csv",
    "/content/comments_reddit_180d_clean_unbalanced.csv",
    "/content/comments_yt_180d_clean_unbalanced.csv",
    "/content/drive/MyDrive/**/*.csv"
]

found = []
for pat in candidates:
    found += glob.glob(pat, recursive=True)

def try_read(path):
    try:
        return pd.read_csv(path)
    except Exception:
        try:
            return pd.read_csv(path, lineterminator="\n", on_bad_lines="skip")
        except Exception:
            return None

dfs = []
for p in found:
    df = try_read(p)
    if df is not None:
        df["__src__"] = os.path.basename(p)
        dfs.append(df)

if not dfs:
    raise FileNotFoundError(
        "No CSVs found. Update the 'candidates' patterns/paths at the top and rerun."
    )
```

```python
# Heuristic: keep only frames with likely comment columns
def is_comment_like(df):
    cols = set(map(str.lower, df.columns))
    return bool(
        {"text","created_dt"}.intersection(cols)
        or {"comment_id","author","platform","show_key"}.issubset(cols)
    )

comment_frames = [d for d in dfs if is_comment_like(d)]

# Try to identify platform by columns or filename
def infer_platform(df):
    cols = set(map(str.lower, df.columns))
    src = str(df.get("__src__", ""))
    if "platform" in cols:
        # trust explicit platform column if present
        val = str(df["platform"].astype(str).str.lower().mode().iloc[0])
        if "reddit" in val: return "reddit"
        if "youtube" in val: return "youtube"
    if re.search(r"reddit", src, re.I): return "reddit"
    if re.search(r"yt|youtube", src, re.I): return "youtube"
    return "unknown"

for d in comment_frames:
    d["platform_inferred"] = infer_platform(d)
```

```python
# --- 2) Harmonize schema
def std_cols(df):
    # make lowercase names
    df = df.copy()
    df.columns = [c.strip() for c in df.columns]
    # unify known columns to standard names if present
    rename_map = {
        "created_at": "created_dt",
        "timestamp": "created_dt",
        "body": "text",
        "comment_text": "text",
        "likecount": "like_count",
        "likes": "like_count",
        "replycount": "reply_count",
        "replies": "reply_count",
        "sc": "sentiment",     # seen "sc" in your PDF – map to sentiment if that's what it is
        "score": "score",      # reddit engagement
        "dominant_topic": "topic",
        "topic_label": "topic"
    }
    for old, new in rename_map.items():
        if old in df.columns and new not in df.columns:
            df = df.rename(columns={old: new})
    # coerces created_dt to datetime if present
    if "created_dt" in df.columns:
        df["created_dt"] = pd.to_datetime(df["created_dt"], errors="coerce", utc=True)
    # era: if not present, infer from 'show_year' or a rule you define
    if "era" not in df.columns:
        # Heuristic: if 'show_year' exists, map years to before/after
        if "show_year" in df.columns:
            # Example cutoff: 2018 and earlier = before; 2019+ = after (adjust as needed)
            df["era"] = np.where(pd.to_numeric(df["show_year"], errors="coerce") >= 2019, "after", "before")
        else:
            # fallback: mark unknown
            df["era"] = "unknown"
    # platform: prefer explicit, else inferred
    if "platform" not in df.columns:
        df["platform"] = df.get("platform_inferred", "unknown")
    # text ensure str
    if "text" in df.columns:
        df["text"] = df["text"].astype(str)
    return df

comment_frames = [std_cols(d) for d in comment_frames]
data = pd.concat(comment_frames, ignore_index=True)

# Keep only eras we care about
data = data[data["era"].isin(["before","after"])].copy()
if data.empty:
    raise ValueError("Could not find rows with era in {'before','after'}. Ensure 'era' or 'show_year' exists or adjust the cutoff rule above.")


```

```python
# --- 0) Install and import dependencies
%pip install -q nltk

import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd
import numpy as np

# --- 1) Download VADER lexicon (once)
nltk.download("vader_lexicon")

# --- 2) Initialize the analyzer
vader = SentimentIntensityAnalyzer()

# --- 3) Define function
def get_sentiment(df):
    df = df.copy()
    if "sentiment" in df.columns and pd.api.types.is_numeric_dtype(df["sentiment"]):
        return df  # already present
    if "text" not in df.columns:
        raise ValueError("No 'text' column to compute sentiment from.")
    df["sentiment"] = df["text"].map(lambda t: vader.polarity_scores(str(t)).get("compound", np.nan))
    return df

# --- 4) Example usage
data = pd.DataFrame({"text": ["I love this concert!", "This was awful.", "It was okay."]})
data = get_sentiment(data)
print(data)
```

```python
# --- 4) Engagement (Reddit: 'score'; YouTube: try typical columns)
def find_engagement_col(df):
    candidates = ["score", "like_count", "reply_count", "favorite_count", "view_count", "comments_count"]
    present = [c for c in candidates if c in df.columns]
    if present:
        # Create a single engagement metric prioritizing typical ones
        priority = {c:i for i,c in enumerate(["score","like_count","reply_count","view_count","favorite_count","comments_count"])}
        present.sort(key=lambda c: priority.get(c, 999))
        return present[0]  # pick the best available
    return None

eng_col = find_engagement_col(data)
if eng_col is None:
    # Create a trivial proxy (length of text) if nothing else exists
    data["engagement_proxy"] = data["text"].str.len()
    eng_col = "engagement_proxy"
data.rename(columns={eng_col: "engagement"}, inplace=True)
```

```python
# --- 5) Topic prevalence
# If you already have a 'topic' column, we'll use it.
# Otherwise, we can build a lightweight BoW+LDA to assign quick topic labels.
# --- 5) Topic prevalence (robust version)
# Uses existing 'topic' if present; otherwise fits a quick BoW + LDA safely.
import numpy as np
import pandas as pd

HAS_TOPIC = "topic" in data.columns

if not HAS_TOPIC:
    # Optional (Notebook): ensure scikit-learn is available
    # %pip -q install scikit-learn

    from sklearn.feature_extraction.text import CountVectorizer
    from sklearn.decomposition import LatentDirichletAllocation

    # --- Prepare texts safely
    if "text" not in data.columns:
        raise ValueError("No 'text' column found for topic modeling.")
    texts = data["text"].fillna("").astype(str).tolist()
    n_docs = len(texts)

    # If no documents, just emit a default topic column and exit
    if n_docs == 0:
        data["topic"] = pd.Series(dtype="Int64")
        HAS_TOPIC = True
    else:
        # --- Start with sensible defaults, auto-adjust if needed
        # For small corpora, make min_df lenient
        base_min_df = 2 if n_docs >= 50 else 1
        base_max_df = 0.9 if n_docs >= 5 else 1.0

        def try_vectorize(min_df, max_df):
            vect = CountVectorizer(lowercase=True, min_df=min_df, max_df=max_df, stop_words="english")
            X = vect.fit_transform(texts)
            return vect, X

        # Attempt 1: user-requested-ish params
        try:
            vect, X = try_vectorize(min_df=base_min_df, max_df=base_max_df)
        except ValueError:
            # Attempt 2: relax thresholds (proportional min_df)
            try:
                vect, X = try_vectorize(min_df=max(0.005, 1.0 / max(n_docs, 1)), max_df=0.95)
            except ValueError:
                # Attempt 3: most permissive
                vect = CountVectorizer(lowercase=True, stop_words="english")
                X = vect.fit_transform(texts)

        # If still empty vocabulary (can happen with ultra-short corpora), fall back to character n-grams
        if X.shape[1] == 0:
            vect = CountVectorizer(analyzer="char", ngram_range=(3, 5))
            X = vect.fit_transform(texts)

        # If still no features, assign a single default topic
        if X.shape[1] == 0:
            data["topic"] = 0
            HAS_TOPIC = True
        else:
            # --- Fit LDA safely
            desired_topics = 8  # your target
            # LDA requires n_components <= number of features and <= number of docs
            k = int(min(desired_topics, X.shape[1], n_docs))
            if k < 1:
                data["topic"] = 0
            else:
                lda = LatentDirichletAllocation(
                    n_components=k,
                    random_state=42,
                    learning_method="batch"
                )
                doc_topic = lda.fit_transform(X)
                data["topic"] = doc_topic.argmax(axis=1).astype(int)

            HAS_TOPIC = True

# After this block:
# - data['topic'] exists and is int-coded (0..k-1)
# - HAS_TOPIC == True
```

```python
# --- 6) Helper: summary tables by era
def era_summary(df, value_col, group_cols=None, roundto=4):
    """
    mean/median/std/count by era; optional extra grouping (e.g., by platform)
    """
    if group_cols is None:
        group_cols = []
    cols = group_cols + ["era"]
    g = df.groupby(cols, dropna=False)[value_col]
    out = g.agg(["mean","median","std","count"]).reset_index()
    return out.round(roundto)

# Re-load and harmonize data to ensure 'era' column exists
comment_frames = [std_cols(d) for d in dfs] # Assuming dfs and std_cols are defined in previous cells
data = pd.concat(comment_frames, ignore_index=True)
data = data[data["era"].isin(["before","after"])].copy() # Ensure 'era' column has valid values

# Re-apply sentiment and engagement calculation (if they were done after initial load)
# Assuming get_sentiment and find_engagement_col are defined in previous cells
data = get_sentiment(data)
eng_col = find_engagement_col(data)
if eng_col is None:
    data["engagement_proxy"] = data["text"].str.len()
    eng_col = "engagement_proxy"
data.rename(columns={eng_col: "engagement"}, inplace=True)


# --- 7) SENTIMENT: compare means (and distribution) before vs after
sent_overall = era_summary(data, "sentiment")
sent_by_platform = era_summary(data, "sentiment", group_cols=["platform"])

# Statistical tests (overall)
before = data.loc[data["era"]=="before","sentiment"].dropna()
after  = data.loc[data["era"]=="after","sentiment"].dropna()

# Welch's t-test (does not assume equal variances)
t_welch = stats.ttest_ind(before, after, equal_var=False)
# Mann-Whitney U (nonparametric, compares distributions)
u_stat, u_p = stats.mannwhitneyu(before, after, alternative="two-sided")

# --- 8) ENGAGEMENT: compare means
eng_overall = era_summary(data, "engagement")
eng_by_platform = era_summary(data, "engagement", group_cols=["platform"])
```

```python
# --- 9) TOPIC PREVALENCE: share of comments per topic by era
if "topic" not in data.columns:
    # If 'topic' column doesn't exist, perform topic modeling
    from sklearn.feature_extraction.text import CountVectorizer
    from sklearn.decomposition import LatentDirichletAllocation

    texts = data["text"].fillna("").tolist()
    # basic tokenization; adjust stop words as needed
    vect = CountVectorizer(lowercase=True, min_df=10, max_df=0.95, stop_words="english")
    X = vect.fit_transform(texts)

    n_topics = 8  # adjust
    lda = LatentDirichletAllocation(n_components=n_topics, random_state=42, learning_method="batch")
    doc_topic = lda.fit_transform(X)
    data["topic"] = doc_topic.argmax(axis=1).astype(int)

if "topic" in data.columns:
    topic_counts = (
        data.groupby(["era","topic"]).size().rename("n").reset_index()
    )
    topic_totals = topic_counts.groupby("era")["n"].transform("sum")
    topic_counts["prevalence"] = (topic_counts["n"] / topic_totals).round(4)
    topic_wide = topic_counts.pivot(index="topic", columns="era", values="prevalence").fillna(0).reset_index()
else:
    topic_counts = pd.DataFrame()
    topic_wide = pd.DataFrame()

# --- 10) Pretty print results
def print_header(s):
    print("\n" + "="*len(s))
    print(s)
    print("="*len(s))

print_header("SENTIMENT — summary by era (overall)")
print(sent_overall.to_string(index=False))

print_header("SENTIMENT — summary by era × platform")
print(sent_by_platform.to_string(index=False))

print_header("SENTIMENT — statistical tests (before vs after)")
print("Welch t-test:      statistic = {:.4f}, p-value = {:.6g}".format(t_welch.statistic, t_welch.pvalue))
print("Mann-Whitney U:    U = {:.0f}, p-value = {:.6g}".format(u_stat, u_p))

print_header("ENGAGEMENT — summary by era (overall)")
print(eng_overall.to_string(index=False))

print_header("ENGAGEMENT — summary by era × platform")
print(eng_by_platform.to_string(index=False))

if "topic" in data.columns:
    print_header("TOPIC PREVALENCE — per era (row = topic id, cols = era, values = share)")
    print(topic_wide.to_string(index=False))
else:
    print_header("TOPIC PREVALENCE")
    print("No topic column found; created none. (Set n_topics or supply a 'topic' column.)")

# --- 11) (Optional) Per-platform statistical tests on sentiment
print_header("SENTIMENT — per-platform tests (before vs after)")
for plat, dfp in data.groupby("platform"):
    b = dfp.loc[dfp["era"]=="before","sentiment"].dropna()
    a = dfp.loc[dfp["era"]=="after","sentiment"].dropna()
    if len(b) > 5 and len(a) > 5:
        t_w = stats.ttest_ind(b, a, equal_var=False)
        U, P = stats.mannwhitneyu(b, a, alternative="two-sided")
        print(f"[{plat}] Welch t: stat={t_w.statistic:.4f}, p={t_w.pvalue:.6g} | M-W U: U={U:.0f}, p={P:.6g}")
    else:
        print(f"[{plat}] Not enough data (n_before={len(b)}, n_after={len(a)})")
```

```python
# --- 12) Save results to CSVs for your report
outdir = "/content"
sent_overall.to_csv(os.path.join(outdir, "sentiment_overall_by_era.csv"), index=False)
sent_by_platform.to_csv(os.path.join(outdir, "sentiment_by_era_by_platform.csv"), index=False)
eng_overall.to_csv(os.path.join(outdir, "engagement_overall_by_era.csv"), index=False)
eng_by_platform.to_csv(os.path.join(outdir, "engagement_by_era_by_platform.csv"), index=False)
if not topic_wide.empty:
    topic_wide.to_csv(os.path.join(outdir, "topic_prevalence_by_era.csv"), index=False)

print("\nSaved: sentiment_overall_by_era.csv, sentiment_by_era_by_platform.csv, "
      "engagement_overall_by_era.csv, engagement_by_era_by_platform.csv, "
      "topic_prevalence_by_era.csv (if topics available).")
```

```python
# ============================================
# Roc Nation Era Visualization Suite
# ============================================

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy import stats

sns.set(style="whitegrid", context="talk", palette="muted")

# --- 1️⃣ Sentiment comparison across eras & platforms
plt.figure(figsize=(9,5))
sns.boxplot(
    data=data, x="era", y="sentiment", hue="platform",
    showmeans=True,
    meanprops={"marker":"o","markerfacecolor":"black","markeredgecolor":"black"}
)
plt.title("Sentiment Distribution: Pre-2019 vs Post-2019 (Roc Nation Era)")
plt.xlabel("Era")
plt.ylabel("Sentiment Score")
plt.legend(title="Platform", loc="best")
plt.tight_layout()
plt.show()

# --- 2️⃣ Mean sentiment (bar with confidence intervals)
plt.figure(figsize=(9,5))
sns.barplot(
    data=data, x="era", y="sentiment", hue="platform",
    ci=95, capsize=.2
)
plt.title("Mean Sentiment by Era and Platform (95% CI)")
plt.xlabel("Era")
plt.ylabel("Mean Sentiment")
plt.tight_layout()
plt.show()

# --- 3️⃣ Comment volume comparison
plt.figure(figsize=(8,5))
vol = data.groupby(["era","platform"]).size().reset_index(name="count")
sns.barplot(data=vol, x="era", y="count", hue="platform")
plt.title("Comment Volume: Pre-2019 vs Post-2019")
plt.ylabel("Number of Comments")
plt.tight_layout()
plt.show()

# --- 4️⃣ Engagement (YouTube likes/views, Reddit scores)
engagement_cols = [c for c in data.columns if c.lower() in ["like_count","view_count","score"]]
if engagement_cols:
    for col in engagement_cols:
        plt.figure(figsize=(8,5))
        sns.violinplot(
            data=data, x="era", y=col, hue="platform", split=True, inner="quart"
        )
        plt.title(f"Engagement ({col}) Distribution: Pre-2019 vs Post-2019")
        plt.ylabel(col.replace("_", " ").title())
        plt.tight_layout()
        plt.show()
else:
    print("No engagement columns found (like_count/view_count/score).")

# --- 5️⃣ Topic prevalence (if topic column exists)
if "topic" in data.columns:
    topic_counts = (
        data.groupby(["era","topic"])
        .size()
        .reset_index(name="count")
    )
    total = topic_counts.groupby("era")["count"].transform("sum")
    topic_counts["percent"] = 100 * topic_counts["count"] / total

    plt.figure(figsize=(10,6))
    sns.barplot(
        data=topic_counts, x="topic", y="percent", hue="era"
    )
    plt.title("Topic Prevalence Before vs After 2019 (Roc Nation)")
    plt.xlabel("Topic")
    plt.ylabel("Share of Comments (%)")
    plt.tight_layout()
    plt.show()

# --- 6️⃣ Statistical test summary (Sentiment differences)
pre2019 = data.query("era=='before'")["sentiment"].dropna()
post2019 = data.query("era=='after'")["sentiment"].dropna()

# Mann-Whitney U test
U, p_u = stats.mannwhitneyu(pre2019, post2019, alternative="two-sided")

# Welch t-test
t_stat, p_t = stats.ttest_ind(pre2019, post2019, equal_var=False)

print("\n📊 Statistical Tests — Sentiment Difference:")
print(f"Welch t-test:  t = {t_stat:.3f},  p = {p_t:.5f}")
print(f"Mann-Whitney U: U = {U:.0f},  p = {p_u:.5f}")

if p_t < 0.05 or p_u < 0.05:
    print("→ Significant difference detected between pre-2019 and post-2019 sentiment.")
else:
    print("→ No statistically significant difference detected.")
```
